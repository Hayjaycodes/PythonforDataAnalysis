```python
#Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt


# #Load data
df= pd.read_csv("data_jobs.csv")

#Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.sample(5)
```

```python
df.index.name= 'job_index'
```

Reset_index( )

```python
df_uk = df[df['job_country'] == 'United Kingdom']
df_uk.reset_index(inplace=True)
df_uk
```

Set Index( )

```python
df_uk.set_index('job_index', inplace=True)
```

```python
df_uk
```

Sort_Index ()

```python
median_pivot = df_uk.pivot_table(values= 'salary_year_avg', index= 'job_title_short', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('job_title_short' ), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)
```
