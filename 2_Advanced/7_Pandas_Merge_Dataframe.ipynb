```python
#Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt


# #Load data
df= pd.read_csv("data_jobs.csv")

#Data Cleanup

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_uk = df[df['job_country'] == 'United Kingdom'].copy()
df_uk
```

```python
df_uk['job_posted_month'] = df_uk['job_posted_date'].dt.strftime('%B')
df_uk
```

```python
df_Uk_pivot= df_uk.pivot_table(index='job_posted_month', columns='job_title_short',aggfunc= 'size')

df_Uk_pivot.reset_index(inplace=True)
df_Uk_pivot['month_no'] = pd.to_datetime(df_Uk_pivot['job_posted_month'], format='%B').dt.month
df_Uk_pivot.sort_values('month_no', inplace=True)
df_Uk_pivot.set_index('job_posted_month', inplace=True)
df_Uk_pivot= df_Uk_pivot.drop(columns='month_no')
df_Uk_pivot
```

```python
df_UK_software_pivot= pd.read_csv("software_jobs.csv", index_col='job_posted_month')

df_UK_software_pivot
```

```python
df_uk_merged = df_Uk_pivot.merge(df_UK_software_pivot, on= 'job_posted_month')
df_uk_merged
```

```python
top_5 = (df_uk_merged.sum()
         .sort_values(ascending=False)
         .head()
         .index. #Get the index of top 5
         to_list())

top_5
```

```python
df_uk_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in the UK')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
