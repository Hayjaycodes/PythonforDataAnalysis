```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  

# Loading Data
df= pd.read_csv("data_jobs.csv")

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

Filter for UK Data Analyst Roles

```python
df_DA_Uk =df[(df['job_country'] == 'United Kingdom') & (df['job_title_short'] == 'Data Analyst')]
df_DA_Uk
```

```python
df_plot= df_DA_Uk['job_location'].value_counts().head(10).to_frame()
sns.set_theme(style= 'ticks')
sns.barplot(  x='count', 
    y='job_location', 
    hue='job_location', 
    data=df_plot, 
    palette='dark:b_r', 
    legend=False)
sns.despine()
plt.title('Counts of Jobs Locations for Data Analysts in the UK')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
# Benefit Analysis
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3)
fig.set_size_inches((12, 5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_Uk[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

# plt.suptitle('Benefit Analysis of Data Jobs', fontsize=16)
plt.show()
```

```python
df_plot= df_DA_Uk['company_name'].value_counts().head(10).to_frame()
sns.set_theme(style= 'ticks')
sns.barplot(  x='count', 
    y='company_name', 
    hue='company_name', 
    data=df_plot, 
    palette='dark:b_r', 
    legend=False)
sns.despine()
plt.title('Counts of Companies for Data Analysts in the UK')
plt.xlabel('Number of Jobs')
plt.ylabel('')
plt.show()
```

```python
df_uk = df[df['job_country'] == 'United Kingdom'].copy()
df_uk
```

```python
df_uk['job_posted_month'] = df_uk['job_posted_date'].dt.strftime('%B')
df_uk
```

```python
df_Uk_pivot= df_uk.pivot_table(index='job_posted_month', columns='job_title_short',aggfunc= 'size')

df_Uk_pivot.reset_index(inplace=True)
df_Uk_pivot['month_no'] = pd.to_datetime(df_Uk_pivot['job_posted_month'], format='%B').dt.month
df_Uk_pivot.sort_values('month_no', inplace=True)
df_Uk_pivot.set_index('job_posted_month', inplace=True)
df_Uk_pivot.drop(columns='month_no')
df_Uk_pivot
```

```python
top_3 = df['job_title_short'].value_counts().nlargest(3).index.tolist()
top_3
```

```python
df_Uk_pivot[top_3].plot(kind='line')
plt.title('Monthly jobs posting for Top 3 job titles in UK')
plt.xlabel('2023')
plt.ylabel('Job count')
plt.legend()
plt.show()
```
